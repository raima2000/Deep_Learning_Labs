```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
SIAnalyzer = SentimentIntensityAnalyzer()
```

```python
SIAnalyzer.polarity_scores(df.iloc[10,-1])
```

```python
df["Polarity"] = df["text"].apply(lambda x: SIAnalyzer.polarity_scores(x))
```

```python
df["Compound_score"] = df["Polarity"].apply(lambda x: x["compound"])
```

```python
df.head()
```

```python
df.tail()
```

```python
def sia_val(x):
    if x < 0:
        return "Negative"
    elif x == 0:
        return "Neutral"
    else:
        return "Positive"
```

```python
df["Sentiment"] = df["Compound_score"].apply(sia_val)
```
